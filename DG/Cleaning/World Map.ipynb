```python
# Import Modules

import json
import pandas as pd
# import numpy as np
# from pandas.io.json import json_normalize
```

```python
# Read json into df

map_df = pd.read_json("../Raw Data/country.json")
map_df.head()
```

```python
# # Get the country name into its own column

map_df["country"] = map_df.properties.map(lambda x: x["name"])
map_df.head()
```

```python
map_df.describe
```

```python
# Read in wine data

wine_df = pd.read_csv("../Cleaned/wine_data.csv")
wine_df.head()
```

```python
wine_df.country.nunique()
```

```python
# Create a dictionarie to rename some countries
rename = {
"US": "United States of America",
"England": "United Kingdom",
"Serbia": "Republic of Serbia"
}

# Replacing the values

wine_df['country'] = wine_df['country'].replace(rename)
wine_df.head()
```

```python
# Remove colums

wine_df = wine_df[['country']]
wine_df.head()
```

```python
# Create groupby object

country_df = wine_df.groupby(['country'])
```

```python
# Total wines

total_wines = country_df['country'].count()
total_wines
```

```python
# Create new df with this data

wine_count_df = pd.DataFrame({
    "total_wines": total_wines
})
wine_count_df.reset_index(level=0, inplace=True)
wine_count_df.head()
```

```python
# Merge wine data and world map data

merge_df = map_df.merge(wine_count_df, how = 'inner', on= 'country')
merge_df.head()
```

```python
# Check amount of countries

merge_df.country.nunique()
```

```python
# Define a function to add the properties back into a dictionary

def add_fields(row):
    properties = row.properties
    return properties.update({
        'country': row.country,
        'total_wines': row.total_wines
    })
```

```python
# Use the function and add to the dataframe

merge_df["merged"] = merge_df.apply(add_fields, axis=1)
merge_df.head()
```

```python
# Export data

merge_df.to_json("../Cleaned/worldmap.json", orient='records')
```

```python
# Export it to a .csv file

merge_df.to_csv("../Cleaned/world_map.csv", index = True)
```
